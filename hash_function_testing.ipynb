Below is the scaffolding for a basic data hasher. We use the DOI as the first step, then the table name, and then the values in that row. 

By having the hash value of a row, we can bring multiple data types together with their confidence intervales, locations, species, etc. and still relate them back together later, and avoid duplication. 

```python
import hashlib as h

def row_hasher(hashfunc=h.sha256, doi="10.1234/a1b2c3d4", table="my_table.csv", values=[1, 2, 3, 4.5, 5, 6]):
    hasher = hashfunc()
    hasher.update(doi.encode("utf8"))
    hasher.update(table.encode("utf8"))
    for value in values:
        hasher.update(str(value).encode('utf8'))
    return hasher.hexdigest()

len(row_hasher(h.md5))
```

```python
def row_hasher_int(hashfunc=h.sha256, doi="10.1234/a1b2c3d4", table="my_table.csv", values=[1, 2, 3, 4.5, 5, 6]):
    hasher = hashfunc()
    hasher.update(doi.encode("utf8"))
    hasher.update(table.encode("utf8"))
    for value in values:
        hasher.update(str(value).encode('utf8'))
    return int(hasher.hexdigest(), 16)

row_hasher_int(h.md5)
```
